```python
from openai import OpenAI
client = OpenAI()

def llm_judge(row):
    prompt = f""""For the user's question, is response 1 better or response 2?
Question: {row.Question}
Response 1: {row['Response3']}
Response 2: {row.Response}

Return either 1 or 2.
"""
    response = client.chat.completions.create(
        model="gpt-4.1",
        messages=[
            {"role": "user", "content": prompt},
        ],
        max_tokens=20,
    )
    return response.choices[0].message.content
```

```python
for _, row in df.iterrows():
    print(llm_judge(row))
```
